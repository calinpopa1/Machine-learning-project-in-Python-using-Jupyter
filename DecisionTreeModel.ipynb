### Définition de méthodes

```python

#Def confusion matrix display
def display_confusion_matrix(confusion_matrix, labels=['1', '2', '3', '4', '5', '6']):
    display = ConfusionMatrixDisplay(confusion_matrix, display_labels = labels)
    display.plot()
    plt.show()
```
